```python
def f(x):
    res  = (1+0.5*x + x*x/12)/(1-0.5*x+x*x/12)
    return res
```

```python
import numpy as np
xshowlist = np.arange(0,4,0.1)
fshowlist = f(xshowlist)
```

```python
import matplotlib.pyplot as plt

plt.plot(xshowlist, fshowlist, label='f')
plt.plot(xshowlist, np.exp(xshowlist), label='e')

plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
