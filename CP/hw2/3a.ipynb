对21个节点使用拉格朗日内插法内插f(x),然后对41个点给出两者的差，并画图显示

用来内插的21个节点

```python
import numpy as np
xlist=np.arange(-1,1.1,0.1)
print(xlist)
```

函数f(x)和21个节点的y值

```python
def f(x):
    return 1/(1+25*x*x)
ylist = f(xlist)
print(ylist)
```

定义拉格朗日多项式L(x)，并定义拉格朗日内插公式

```python
def laglangr(x,j):
        res = 1
        for m in range(21):
            if(j==m):
                continue
            res *= (x-xlist[m])/(xlist[j]-xlist[m])
        return res

def Pn(x):
    res = 0
    for i in range(21):
        res += ylist[i] * laglangr(x, i)
    return res
```

用Pn计算21个节点的y值

```python
Pylist = Pn(xlist)
print(Pylist)
```

构建41个点，然后分别用f(x)和Pn(x)计算y值，并给出差

```python
xshowlist = np.arange(-1,1.05,0.05)
fshowlist = f(xshowlist)
pshowlist = Pn(xshowlist)
print(fshowlist)
print(pshowlist)
print(pshowlist-fshowlist)
```

画图比较

```python
import matplotlib.pyplot as plt

plt.plot(xshowlist, fshowlist, label='f')
plt.plot(xshowlist, pshowlist, label='Pn')

plt.legend()

plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
