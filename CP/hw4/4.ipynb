设置步长，三个参数，三个初始值，在for循环里每次计算一个步长，然后将x，y，z的轨迹画在三维图上

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

nstep = 100000
xfinal = 10.0
xini = 0.0
beta = 8.0 / 1.0
rho = 28.0
sigma = 5
x = np.zeros(nstep + 1)
y = np.zeros(nstep + 1)
z = np.zeros(nstep + 1)
x[0] = 12.0
y[0] = 4.0
z[0] = 0.0
xstep = (xfinal - xini) / nstep

for i in range(nstep):
    x[i + 1] = x[i] + xstep * (-beta * x[i] + y[i] * z[i])
    y[i + 1] = y[i] + xstep * (-sigma * y[i] + sigma * z[i])
    z[i + 1] = z[i] + xstep * (-y[i] * x[i] + rho * y[i] - z[i])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
